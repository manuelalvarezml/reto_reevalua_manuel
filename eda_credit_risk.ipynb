```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("data_files/credit_risk_with_targets_cleaned_final.csv")

print("\nüîç Basic Info:")
print(df.info())

print("\nüìâ Missing Values:")
print(df.isnull().sum())

print("\nüìä Target Distribution:")
print(df['target'].value_counts(normalize=True).rename("proportion"))

print("\nüìê Numerical Summary:")
print(df.describe())

print("\nüî§ Categorical Value Counts:")
categorical_cols = ['Sex', 'Job', 'Housing', 'Saving accounts', 'Checking account', 'Purpose']
for col in categorical_cols:
    print(f"\nüß© {col}:")
    print(df[col].value_counts(dropna=False))

# Plot distributions
print("\nüìà Plotting distributions...")

# Set style
sns.set(style="whitegrid")

# Histogram: Age
plt.figure(figsize=(6, 4))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title("Age Distribution")
plt.show()

# Count plot: Job
plt.figure(figsize=(6, 4))
sns.countplot(x='Job', data=df)
plt.title("Job Type Count")
plt.show()

# Count plot: Housing
plt.figure(figsize=(6, 4))
sns.countplot(x='Housing', data=df)
plt.title("Housing Type Count")
plt.show()

# Target vs Numerical Features
plt.figure(figsize=(6, 4))
sns.boxplot(x='target', y='Credit amount', data=df)
plt.title("Credit Amount by Risk")
plt.show()

plt.figure(figsize=(6, 4))
sns.boxplot(x='target', y='Duration', data=df)
plt.title("Duration by Risk")
plt.show()
```
